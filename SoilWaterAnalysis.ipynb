##Bring in libraries and dataframes and set indexes

```python
import datetime
import pandas as pd
import numpy as np
#import ggplot as gg
import matplotlib.pyplot as plt
%matplotlib inline

#Read in data
AllData=pd.read_csv('K:\Rainshelter\Barley_2014_15\LoggedData\RainShelterAlpha_CS650.dat', #specify file path for data to read in
                         parse_dates=True, #tell the function to parse date columns to datetime formats
                         dayfirst=True, #tell the function that the day is before the year in the data i.e format='%d/%m/%Y %H:%M'
                         skiprows = [0,2,3], #leave out rows 1, 3 and 4 which have redundant information
                         index_col = 0, #Use the first column, which is Date, as an index
                         na_values = 'NAN')

#Bring in index data
AllDataIndex=pd.read_csv('C:\GitHubRepos\RainShelterBarley\SoilWaterAndTempIndex.csv',
                         index_col = 0)
#Apply indexes to data
AllDataTransposed = AllData.transpose()
AllDataIndexed = pd.concat([AllDataIndex,AllDataTransposed], axis=1)
AllDataIndexed.index.name='ColumnHeader'
AllDataIndexed.set_index(['Measurement','Cultivar','Irrigation','Treatment','Sensor', 'MUX', 'Port', 'Depth','Units','Summary','Plot','Block'], 
                        append=False, inplace=True)
AllDataIndex.sort_index()
Data=AllDataIndexed.transpose()
#Data.sort(inplace=True)

#Calculate treatment means
DataMeans = Data.groupby(level=['Measurement','Depth','Irrigation', 'Cultivar'],axis=1).mean()
```

## Set time slice to graph

```python
from datetime import timedelta #Import function needed for doing date deltas 
EndDate = Data.index.max(axis=0)  #Set end of series as most recent date in file
EndDateString = EndDate.strftime("%Y-%m-%d")  #Turn this to a string
PlotDuration = 2                              #Set the duration that you want to graph for
StartDate = EndDate - timedelta(weeks=PlotDuration)       #Set start date the specified duration before the end date 
StartDateString = StartDate.strftime("%Y-%m-%d")           #Turn that into a string
```

##Graph all sensors to check for anomolies

```python
DataLookSlice = Data.ix[StartDateString:]

Fig = plt.figure(figsize=(18, 30));
    
def MakePlot(Variable, Position, Unit):
        Fig.add_subplot(Position)
        plt.title(Variable, fontsize=28);
        plt.plot(DataLookSlice.index,DataLookSlice.ix[StartDateString:,[X for X in DataLookSlice.columns if Variable in X]]);
        plt.ylabel(Unit, fontsize=22);
        plt.tick_params(labelsize=16);
        return;
MakePlot('VolumetricWaterContent',311,'mm/mm')
MakePlot('Temperature',312,'oC')
MakePlot('ElectricalConductivity',313,'dS/m')
```

```python
DataMeans = Data.groupby(level=['Measurement', 'Depth', 'Irrigation', 'Cultivar'],axis=1).mean()


#Calculate the water content of the soil profile by multiplying the volumetric water content by each layers
#depth and summing.  The 0-15 layers are divided by 2 to average the two readings
ProfileWater = DataMeans.VolumetricWaterContent.ix[:,'0-15(InRow)'] * 150/2 + \
               DataMeans.VolumetricWaterContent.ix[:,'0-15(btwRow)'] * 150/2 + \
               DataMeans.VolumetricWaterContent.ix[:,'15-30'] * 150 + \
               DataMeans.VolumetricWaterContent.ix[:,'30-60'] * 150 + \
               DataMeans.VolumetricWaterContent.ix[:,'60-90'] * 150 + \
               DataMeans.VolumetricWaterContent.ix[:,'90-120'] * 150 + \
               DataMeans.VolumetricWaterContent.ix[:,'120-150'] * 150 + \
               DataMeans.VolumetricWaterContent.ix[:,'150-180'] * 150
ProfileWater.plot(figsize=(18,18))
plt.ylabel('Profile Water (mm)', fontsize=30);
plt.tick_params(labelsize=25)
plt.title('RainShelter Barley 2014/15', fontsize=28);
```

##Graph treatment means by layer

```python
#DataMeans = Data.groupby(level=['Measurement', 'Depth', 'Irrigation', 'Cultivar'],axis=1).mean()
MeanLookSlice = DataMeans.ix[StartDateString:]

Fig = plt.figure(figsize=(18, 60))

def MakePlot(Position, Depth, LayerThickness, Ymax):
        Fig.add_subplot(8,1,Position)
        plt.title(Depth + '  VolumetricWaterContent', fontsize=28);
        plt.plot(MeanLookSlice.index,MeanLookSlice.VolumetricWaterContent.ix[StartDateString:,Depth]*LayerThickness);
        plt.ylabel('mm/mm', fontsize=22);
        plt.tick_params(labelsize=16);
        plt.ylim(20,Ymax);
        return;

MakePlot(1,'0-15(InRow)',150,50)
MakePlot(2,'0-15(btwRow)',150,50)
MakePlot(3,'15-30',150,70)
MakePlot(4,'30-60',300,140)
MakePlot(5,'60-90',300,140)
MakePlot(6,'90-120',300,140)
MakePlot(7,'120-150',300,140)
MakePlot(8,'150-180',300,140)
```
