# Water use and energy balance of two barley cultivars of contrasting canopy structure under diffent irrigation treatments

This notebook present some live graphs showing the data that is being collected for a trial being conducted in the rainshelter
* Two varieties of barley sown on 15th of October 2014:
    * 'Dash' is a modern variety with errect leaves and awns
    * 'Omaka' is an old variety with prostrate leaves and awns
* Three irrigation treatments are being applied:
    * 'High' replacing crop water use weekly
    * 'Mid' replacing 1/2 crop water use monthly
    * 'Low' a little bit of water every so often.
* Four replicates of each treatment giving 24 plots
* Campbell CS650 sensors installed in each plot
    * measure:
        * Volumetric water content over the length of their 30cm rods
        * Temperature in the head of the sensor
        * Electrical conductivity of the soil over the length of their 30cm rods
        * Correct water content for temperature and electical conductivity variation
    * installed at 8 positions in each plot
        * 2 sensors installed sampling 0-15cm depth
        * 1 sensor installed at 15-30, 30-60, 60-90, 90-120, 120-150 and 150-180 cm depth
        * 24 plots in the trial so 192 sensor are being recorded.
* Apogee infra red radiometers have been installed over every plot logging surface temperature
* Campbell T108 temperature sensors installed in the soil (just burried) in every plot to recored soil surface temperature

12 of the plots (3 replicates of the high and low irrigation treatments) have also been instrumented with:
* Solams linear PAR sensors to record radiation interception
* Hukseflux heat flux plates to record soil heat flux below 15 cm depth
* Averaging thermo couples to record soil temperature in the 15 cm above the heat flux plates
* Campbell pyranometers pointing downward to measure reflected radiation

Below are some the the results as they roll in.

##Bring in libraries and dataframes and set indexes

```python
import datetime
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Read in data
AllData=pd.read_csv('K:\Rainshelter\Barley_2014_15\LoggedData\RainShelterAlpha_CS650.dat', #specify file path for data to read in
                         parse_dates=True, #tell the function to parse date columns to datetime formats
                         dayfirst=True, #tell the function that the day is before the year in the data i.e format='%d/%m/%Y %H:%M'
                         skiprows = [0,2,3], #leave out rows 1, 3 and 4 which have redundant information
                         index_col = 0, #Use the first column, which is Date, as an index
                         na_values = 'NAN')

#Bring in index data
AllDataIndex=pd.read_csv('C:\GitHubRepos\RainShelterBarley\SoilWaterAndTempIndex.csv',
                         index_col = 0)
#Apply indexes to data
AllDataTransposed = AllData.transpose()
AllDataIndexed = pd.concat([AllDataIndex,AllDataTransposed], axis=1)
AllDataIndexed.index.name='ColumnHeader'
AllDataIndexed.set_index(['Measurement','Depth','Cultivar','Irrigation','Plot','Sensor', 'MUX', 'Port','Units','Summary','Block','Treatment'], 
                        append=False, inplace=True)
AllDataIndexed.sort(inplace=True)
Data=AllDataIndexed.transpose()
Data.index = Data.index.to_datetime()  ## for some reason the concat function changes the data type on the date indes so need to change it back
```

Calculate treatment means omitting data prior to 2014-11-05 08:00:00 to avoid NaN values

```python
DataMeans =  Data.ix['2014-11-05 08:00:00':].groupby(level=['Measurement','Depth','Irrigation', 'Cultivar'],axis=1).mean()
DataMeans =  DataMeans.dropna(axis=1) #For some reason it keeps non valid combinations in so need to extract with this function
```

##Graph profile soil water deficits

Calculate the water content of the soil profile by multiplying the volumetric water content by each layers
depth and summing.  The 0-15 layers are divided by 2 to average the two readings

```python
ProfileWater = DataMeans.VolumetricWaterContent.ix[:,'D1I'] * 150/2 + \
               DataMeans.VolumetricWaterContent.ix[:,'D1B'] * 150/2 + \
               DataMeans.VolumetricWaterContent.ix[:,'D2'] * 150 + \
               DataMeans.VolumetricWaterContent.ix[:,'D3'] * 300 + \
               DataMeans.VolumetricWaterContent.ix[:,'D4'] * 300 + \
               DataMeans.VolumetricWaterContent.ix[:,'D5'] * 300 + \
               DataMeans.VolumetricWaterContent.ix[:,'D6'] * 300 + \
               DataMeans.VolumetricWaterContent.ix[:,'D7'] * 300
                
FieldCapacity = ProfileWater.resample('D', how='max')
FieldCapacity = FieldCapacity.ix['2014-11-05'] + 40
SoilWaterDeficit = -(FieldCapacity - ProfileWater)
SoilWaterDeficit.plot(figsize=(18,18),style=['b-','b--','r-','r--','g-','g--'])
plt.ylabel('Soil Water Deficit (mm)', fontsize=30);
plt.tick_params(labelsize=25)
plt.title('RainShelter Barley 2014/15', fontsize=28);
plt.legend(loc='lower left',fontsize=20)
```

## Set time slice to graph

```python
from datetime import timedelta #Import function needed for doing date deltas 
EndDate = Data.index.max(axis=0)  #Set end of series as most recent date in file
EndDateString = EndDate.strftime("%Y-%m-%d")  #Turn this to a string
PlotDuration = 1                              #Set the duration that you want to graph for
StartDate = EndDate - timedelta(weeks=PlotDuration)       #Set start date the specified duration before the end date 
StartDateString = StartDate.strftime("%Y-%m-%d")           #Turn that into a string
```
