##Bring in libraries and dataframes and set indexes

```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Read in data
AllData=pd.read_csv('K:\Rainshelter\Barley_2014_15\LoggedData\RainShelterBeta_Energy.dat', #specify file path for data to read in
                         parse_dates=True, #tell the function to parse date columns to datetime formats
                         dayfirst=True, #tell the function that the day is before the year in the data i.e format='%d/%m/%Y %H:%M'
                         skiprows = [0,2,3], #leave out rows 1, 3 and 4 which have redundant information
                         index_col = 0, #Use the first column, which is Date, as an index
                         na_values = 'NAN')
#Bring in index data
AllDataIndex=pd.read_csv('C:\GitHubRepos\RainShelterBarley\RadiationAndTempIndex.csv',
                         index_col = 0)

#Apply indexes to data
AllDataTransposed = AllData.transpose()
AllDataIndexed = pd.concat([AllDataIndex,AllDataTransposed], axis=1)
AllDataIndexed.index.name='ColumnHeader'
AllDataIndexed.set_index(['Measurement','Cultivar','Irrigation','Treatment','Units','Summary','Plot','Column','Block'], 
                         append=False, inplace=True)
Data=AllDataIndexed.transpose()
Data.sort(inplace=True)

#Calculate treatment means
DataMeans = Data.groupby(level=['Measurement','Irrigation', 'Cultivar'],axis=1).mean()
```

## Set time slice to graph

```python
from datetime import timedelta #Import function needed for doing date deltas 
EndDate = Data.index.max(axis=0)  #Set end of series as most recent date in file
EndDateString = EndDate.strftime("%Y-%m-%d")  #Turn this to a string
PlotDuration = 1                              #Set the duration that you want to graph for
StartDate = EndDate - timedelta(weeks=PlotDuration)       #Set start date the specified duration before the end date 
StartDateString = StartDate.strftime("%Y-%m-%d")           #Turn that into a string
```

##Graph met data the check

```python
DataMeans.ix[StartDateString:,['BatteryVoltage','AirTemperature','IncomingRadnKW','Leafwetness','PannelTemperature','Rainfall','RelativeHumidity','Windspeed']].plot(subplots=True, figsize=(18, 60),
                    style='r-',
                    linewidth = 2)                                   
```

##Graph all sensors to check for anomolies

```python
Data.ix[StartDateString:,[X for X in Data.columns if 'HeatFlux' in X]].plot(style = ['r-','b--','b--','r-','r--','r--','b-','b--','r-','b-','r--','b-'], figsize=(18,10), legend = False)
Data.ix[StartDateString:,[X for X in Data.columns if 'IR_SurfaceTemp' in X]].plot(legend=False, figsize=(18,10))
Data.ix[StartDateString:,[X for X in Data.columns if 'IR_SensorBodyTemp' in X]].plot(legend=False, figsize=(18,10))
Data.ix[StartDateString:,[X for X in Data.columns if 'SoilTemp15cm' in X]].plot(style = ['r-','b--','b--','r-','r--','r--','b-','r-','b--','b-','r--','b-'],legend=False, figsize=(18,10))
Data.ix[StartDateString:,[X for X in Data.columns if 'BelowCanopyPAR' in X]].plot(style = ['r-','b--','b--','r-','r--','r--','b-','b--','r-','b-','r--','b-'],legend=False, figsize=(18,10))
Data.ix[StartDateString:,[X for X in Data.columns if 'SurfaceSoilTemp' in X]].plot(legend=False, figsize=(18,10))
```

##Graph treatment means

```python
DataMeans = Data.groupby(level=['Measurement','Irrigation', 'Cultivar'],axis=1).mean()
DataMeans.ix[StartDateString:,['IncomingRadnKW','ReflectedRadnKW']].plot(figsize=(18,10))
DataMeans.ix[StartDateString:,['PannelTemperature','AirTemperature','SurfaceSoilTemp','SoilTemp15cm']].plot(figsize=(18,10))
DataMeans.ix[StartDateString:,['IR_SurfaceTemp','SurfaceSoilTemp']].plot(figsize=(18,10))
DataMeans.ix[StartDateString:,['AboveCanopyPAR','BelowCanopyPAR']].plot(figsize=(18,10))
DataMeans.ix[StartDateString:,'HeatFlux'].plot(figsize=(18,10))
```

##Calculate and graph radiation interception and reflection

```python
FractPARint = DataMeans.ix[StartDateString:,'BelowCanopyPAR'] / DataMeans.ix[StartDateString:,0] 
FractPARint.plot(figsize=(18,10),
                 ylim=(0,1))
FractRadRefl = DataMeans.ix[StartDateString:,'ReflectedRadnKW'] / Data.ix[StartDateString:,84] 
FractRadRefl.plot(figsize=(18,10),
                 ylim=(0,.4))
```
