```python
CanopyTempDiffs = CanopyTempData
CanopyTempDiffs.ix[:,0:23] = CanopyTempDiffs.ix[:,0:23] - CanopyTempData.ix[:,'AirT']
```

```python
DaytimeMeanCanTempDiffs = CanopyTempDiffs.ix[CanopyTempData['Dailylight'] == 'daylight',:]
DaytimeMeanCanTempDiffs = CanopyTempDiffs.resample('D', how='mean')
```

```python
DaytimeMeanCanTempDiffs.ix['2013-12-15':'2014-2-20',
                     ['Lucerne']].plot(figsize=(14, 10),
                     color=('b','y','g','r','g','b','r','y','r','g','g','r','b','b'),
                     ylim=(-5,10), 
                     legend = False)
```

```python
CoverTempDiff = DaytimeMeanCanTempDiffs.groupby(axis = 1, level=['Irrigation','Species']).mean()
```
